```python
with open('./Assets/8.txt') as f:
    contents = f.read()

contents = contents.split("\n")

input = []
for content in contents:
  input.append([int(i) for i in content])

print(input)
```

```python
def checkBoundaries(x, y, l, b):

  right = True
  left = True
  top = True
  bottom = True

  for i in range(x+1, l):
    if input[x][y] <= input[i][y]:
      right = False
      break
  
  for i in range(x-1, -1, -1):
    if input[x][y] <= input[i][y]:
      left = False
      break

  for i in range(y+1, b):
    if input[x][y] <= input[x][i]:
      bottom = False
      break
  
  for i in range(y-1, -1, -1):
    if input[x][y] <= input[x][i]:
      top = False
      break
  
  return top or bottom or left or right
```

```python
length = len(input)
breadth = len(input[0])

count = 0

for i in range(length):
  for j in range(breadth):
    if i == 0 or j == 0 or i == length-1 or j == breadth-1:
      count += 1
      # print(i, j, input[i][j])
    elif checkBoundaries(i, j, length, breadth):
      count += 1
      # print(i, j, input[i][j], True)


print(count)
```

```python
def calculateScenicScore(x, y, l, b):

  right = 0
  left = 0
  top = 0
  bottom = 0

  for i in range(x+1, l):
    right += 1
    if input[x][y] <= input[i][y]:
      break


  for i in range(x-1, -1, -1):
    left += 1
    if input[x][y] <= input[i][y]:
      break

  for i in range(y+1, b):
    bottom += 1
    if input[x][y] <= input[x][i]:
      break

  for i in range(y-1, -1, -1):
    top += 1
    if input[x][y] <= input[x][i]:
      break

  
  return top * bottom * left * right
```

```python
length = len(input)
breadth = len(input[0])

# print(length)
# print(breadth)

max_score = 0

for i in range(length):
  for j in range(breadth):
    score = calculateScenicScore(i, j, length, breadth)
    if(max_score < score):
      # print(i, j, input[i][j], score, max_score)
      max_score = score

print(max_score)
```
