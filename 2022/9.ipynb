```python
with open('./Assets/test.txt') as f:
  contents = f.read()

contents = contents.split("\n")

input = []
for content in contents:
  input.append([content.split(' ')[0], int(content.split(' ')[1])])

print(input)
```

```python
import math


def moveTail(new_head_pos, head_pos, tail_pos, dir):
  diff = math.sqrt((new_head_pos[0] - tail_pos[0])
                   ** 2 + (new_head_pos[1] - tail_pos[1])**2)
  # print(diff)

  if diff == 2.0:
    if dir == 'R':
      tail_pos[1] = new_head_pos[1]-1
    elif dir == 'L':
      tail_pos[1] = new_head_pos[1]+1
    elif dir == 'U':
      tail_pos[0] = new_head_pos[0]-1
    elif dir == 'D':
      tail_pos[0] = new_head_pos[0]+1
  elif diff > 2.0:
    if dir == 'R':
      tail_pos[1] = new_head_pos[1]-1
      tail_pos[0] = new_head_pos[0]
    elif dir == 'L':
      tail_pos[1] = new_head_pos[1]+1
      tail_pos[0] = new_head_pos[0]
    elif dir == 'U':
      tail_pos[0] = new_head_pos[0]-1
      tail_pos[1] = new_head_pos[1]
    elif dir == 'D':
      tail_pos[0] = new_head_pos[0]+1
      tail_pos[1] = new_head_pos[1]
  
  return tail_pos

def moveHead(dir, pos, head_pos, tail_pos):
  new_head_pos = [head_pos[0], head_pos[1]]
  
  if dir == 'R':
    new_head_pos[1] += pos
  elif dir == 'L':
    new_head_pos[1] -= pos
  elif dir == 'U':
    new_head_pos[0] += pos
  elif dir == 'D':
    new_head_pos[0] -= pos


  tail_pos = moveTail(new_head_pos, head_pos, tail_pos, dir)
  
  return new_head_pos, tail_pos
```

```python
head_pos = [0,0]
tail_pos = [0,0]

tail_positions = []
for rule in input:
  print("move: " + rule[0] + " pos: " + str(rule[1]))
  for _ in range(rule[1]):
    head_pos, tail_pos = moveHead(rule[0], 1, head_pos, tail_pos)
    print(head_pos, tail_pos)
    tail_positions.append(str(tail_pos))


```
