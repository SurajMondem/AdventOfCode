```python
with open('./Assets/4.txt') as f:
    contents = f.read()

contents = contents.split("\n")

input = []
for content in contents:
  input.append([[j for j in i.split('-')] for i in content.split(',')])

print(input)
```

```python
def fullyContained(elf1, elf2):
  if ((int(elf1[0]) <= int(elf2[0]) and int(elf1[1]) >= int(elf2[1])) or (int(elf2[0]) <= int(elf1[0]) and int(elf2[1]) >= int(elf1[1]))):
    print(elf1, elf2)
    return True
  else:
    return False
```

```python
count = 0
for group in input:
  if fullyContained(group[0], group[1]):
    count+=1
  
print(count)
```

```python
def fullyContained2(elf1, elf2):
  if (int(elf1[0]) > int(elf2[1]) or int(elf1[1]) < int(elf2[0])):
    print(elf1, elf2)
    return True
  else:
    return False
```

```python
count = 0
for group in input:
  if fullyContained2(group[0], group[1]):
    count += 1

print(len(input)-count)
```
