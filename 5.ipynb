```python
import math

with open('./Assets/5.txt') as f:
    contents = f.read()

stacks, steps = contents.split("\n\n")

print(stacks)

converted_stack = stacks.replace('[', ' ').replace(
    ']', ' ').replace('\n', '*').replace(' ', '+')

# print(converted_stack)

count = 0
for key in stacks.split():
  if key.isnumeric():
    count +=1


fr = []
sc = []
th = []
all_stacks = [[] for _ in range(count)]

tab = 4
c = 0
for ch in converted_stack:
  if ch == '+':
    c += 1
  elif ch.isnumeric():
    break
  elif ch.isalpha():
    if c == 0 or (c-1)/tab == 0:
      all_stacks[0].append(ch)
      c += 1
    elif c/tab == 1 or (c-1)/tab== 1:
      all_stacks[1].append(ch)
      c += 1
    elif c/tab == 2 or (c-1)/tab == 2:
      all_stacks[2].append(ch)
      c += 1
    elif c/tab == 3 or (c-1)/tab == 3:
      all_stacks[3].append(ch)
      c += 1
    elif c/tab == 4 or (c-1)/tab == 4:
      all_stacks[4].append(ch)
      c += 1
    elif c/tab == 5 or (c-1)/tab == 5:
      all_stacks[5].append(ch)
      c += 1
    elif c/tab == 6 or (c-1)/tab == 6:
      all_stacks[6].append(ch)
      c += 1
    elif c/tab == 7 or (c-1)/tab == 7:
      all_stacks[7].append(ch)
      c += 1
    elif c/tab == 8 or (c-1)/tab == 8:
      all_stacks[8].append(ch)
      c += 1
  elif ch == '*':
    c = 0


# print(all_stacks)

all_stacks = [list(reversed(x)) for x in all_stacks]
print(all_stacks)
```

```python
raw_instructions = [x.split() for x in steps.split('\n')]
# print(raw_instructions)

instructions = []
for ins in raw_instructions:
  arr = []
  for i in ins:
    if i.isnumeric():
      arr.append(int(i))
  instructions.append(arr)

print(instructions)
```
